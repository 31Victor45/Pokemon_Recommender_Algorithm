# Limpieza de los datos

En este jupiter nos centraremos en obtener los pokemon del dataset desde primera a cuarta generacion y tambien nos aceguraremos que no haya mega evoluciones y que solo se incluya un pokemon original (es decir no aseptare variaciones o otras formas que pueda tener un pokemon, nos quedaremos con la forma base del mismo)

```python
# Importamos las librerias necesarias
import pandas as pd
import numpy as np
```

```python
# Trabajaremos con pokemons de la 1ra a las 4ta generacion
df_pkmn = pd.read_csv("pkmn.csv")
df_pkmn = df_pkmn.iloc[:553, :]
df_pkmn.shape
```

```python
# Filtramos los pokemon de primera generacion
df_pkmn_gen1 = df_pkmn[((df_pkmn["Generation"] == 1) & (~df_pkmn["Name"].str.contains(" "))) | (df_pkmn["Name"] == "Mr. Mime")]
df_pkmn_gen1
```

```python
# Filtramos los pokemon de segunda generacion
df_pkmn_gen2 = df_pkmn[(df_pkmn["Generation"] == 2) & (~df_pkmn['Name'].str.contains(' '))]
df_pkmn_gen2
```

```python
# Filtramos los pokemon de tercera generacion
df_pkmn_gen3 = df_pkmn[((df_pkmn["Generation"] == 3) & (~df_pkmn['Name'].str.contains(' '))) | (df_pkmn["Name"] == "DeoxysNormal Forme")]
df_pkmn_gen3
```

```python
# Filtramos los pokemon de cuarta generacion
df_pkmn_gen4 = df_pkmn[((df_pkmn["Generation"] == 4) & (~df_pkmn['Name'].str.contains(' '))) | (df_pkmn["Name"] == "WormadamPlant Cloak") | (df_pkmn["Name"] == "Mime Jr.")| (df_pkmn["Name"] == "GiratinaAltered Forme")| (df_pkmn["Name"] == "ShayminLand Forme")]
df_pkmn_gen4
```

```python
# Unimos en un dataframe las 4 generaciones pokemon filtradas
df_pkmn_work = pd.concat([df_pkmn_gen1, df_pkmn_gen2, df_pkmn_gen3, df_pkmn_gen4], ignore_index=True)
df_pkmn_work
```

Ahora tomare el dataset ya modificado y y quitare la columna "Total" ya que esta no serÃ¡ util para el analisis exporatorio de datos que planeo hacer

No obstante quiero mencionar que el sistema de recomendacion que planeo crear no tendra en cuenta la generacion a la que pertenezca el pokemon y si el mismo es legendario o no ya que si bien esta informacion seria util de cara al analisis exploratorio, no la encuetro practica a la hora de crear el sistema de recomendacion

```python
df_pkmn_work.drop(columns=['Total']).to_csv('pkmn_exploratory.csv', index=False)
```

Ahora crearemos un csv que tomarÃ¡ los pokemon de las 4 generaciones antes filtradas y eliminaremos variables que no concideramos utiles para el sistema de recomendacion que queremos crear

```python
df_pkmn_work.drop(columns=['Generation', 'Legendary', 'Total']).to_csv('df_pkmn_clean.csv', index=False)
```
